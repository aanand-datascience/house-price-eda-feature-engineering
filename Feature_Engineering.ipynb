## 𝗠𝗶𝘀𝘀𝗶𝗻𝗴 𝗩𝗮𝗹𝘂𝗲𝘀

```python
##Let us capture all the non values
##First Lets handle Categorical feature which are missing 

########### this proces is called list comprehension ###########

features_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes=='object']

for feature in features_nan:
    print("{}: {}% missing values".format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

✅ If you call function without assigning
replace_cat_feature(dataset, features_nan)


Function runs…
Modified dataset is created…
But you don’t save it anywhere 😅

So Python discards it.

Original dataset remains unchanged.

```python
##Replce missing values with a new Label
def replace_cat_feature(dataset,features_nan):
    data=dataset.copy()
    data[features_nan]=data[features_nan].fillna('Missing')
    return data
dataset=replace_cat_feature(dataset,features_nan)

dataset[features_nan].isnull().sum()
```

```python
dataset.head()
```

```python
dataset['Alley']
```

```python
## now Lets check for numerical variables the contains missing values
numerical_with_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes!='object']

## We will print the numerical nan variable and percentage of missing values

for feature in numerical_with_nan:
    print("{}: {}% missing values".format(feature,np.around(dataset[feature].isnull().mean(),4)))
```

```python
## To get all the missing index
dataset[dataset['LotFrontage'].isnull()].index
```

```python
dataset['LotFrontage'].head(50)
```

```python
##Replacing the numerical missing  values

for feature in numerical_with_nan:
    ## we will replace by using median Since there are outliers
    median_values=dataset[feature].median()
    
    ## Create a new feature to capture nan values.
    dataset[feature+'nan']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_values, inplace= True)
    
dataset[numerical_with_nan].isnull().sum()
```

```python
dataset.head(50)
```

```python
## Temporal Variable (Date Time Variables )
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    dataset[feature]=dataset['YrSold']-dataset[feature]
```

## 𝗙𝗲𝗮𝘁𝘂𝗿𝗲 𝗘𝗻𝗴𝗶𝗻𝗲𝗲𝗿𝗶𝗻𝗴 𝗣𝗮𝗿𝘁 - 𝟮

Numerical Variables

Since the numerical variables are skewed we will perform log normal distribution

```python
dataset.head()
```

✅ Why we do this (main reasons)
1️⃣ Reduce skewness

Many real-world numeric features are right-skewed (few very large values).

Log makes distribution more balanced (closer to normal).

```python
import numpy as np
num_features=['LotFrontage','LotArea','1stFlrSF','GrLivArea','SalePrice']

for feature in num_features:
    dataset[feature]=np.log(dataset[feature])
```

```python
dataset.head()
```

Handling Rare Categorical Feature

---> We will remove categorical variables that are present leass than 1% of observations

```python
categorical_features=[feature for feature in dataset.columns if dataset[feature].dtype=='object']
```

```python
categorical_features
```

```python
for feature in categorical_features:
    temp=dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
```

```python
dataset.head(60)
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

Feature Scaling

1. StandardScalar
2. MinMaxScalar

👉 Feature scaling is used to bring all features to the same range so machine learning models can learn efficiently and fairly.

* StandardScaler → handles better

* MinMaxScaler → compresses everything because max becomes huge

✅ 1️⃣ StandardScaler (Z-score scaling) 
Values are centered around 0.
Range is not fixed (can be −3, 2, etc.).

✅ 2️⃣ MinMaxScaler (Normalization)
Rescales values into a fixed range, usually:

0
 to 
1


```python
feature_scale=[feature for feature in dataset.columns if feature not in ['Id','SalePrice']]

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(dataset[feature_scale])
```

👉 We scale features so that all variables have equal influence on the model and training becomes faster and more stable.

⭐ Transform kya hai?

Scaling rule ko data pe apply karna.

```python

scaler.transform(dataset[feature_scale])
```

💬 One line summary

👉 Scale features separately → convert to DataFrame → join with Id & target → new dataset ready.

```python
# transform the train and test set, and add on the Id and SalePrice variables
data = pd.concat([dataset[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(dataset[feature_scale]), columns=feature_scale)],
                    axis=1)
```
